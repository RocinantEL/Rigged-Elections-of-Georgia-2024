```python
import random
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import normaltest
import time
```

```python
#File reading and procesing for use
##############################################################################################################
# 5% bin size data ("CEC" official data)
file_pathreal = r'C:\Users\user\Desktop\Everything\Projects\FermiLab\realdist5%.csv'
ddff = pd.read_csv(file_pathreal)
percent=ddff['Percent'][0:20]
freq=ddff['Frequency'][0:20]
##############################################################################################################
# 5% bin size data (comp simulation)
file_hist=r'C:\Users\user\Desktop\Everything\Projects\FermiLab\hist.csv'
dfsigma=pd.read_csv(file_hist)
##############################################################################################################
# 1% bin size data (comp simulation)
file_hist1=r'C:\Users\user\Desktop\Everything\Projects\FermiLab\hist1.csv'
dfhist1=pd.read_csv(file_hist1)
##############################################################################################################
# 1% bin size data ("CEC" official data)
file_pathrea2 = r'C:\Users\user\Desktop\Everything\Projects\FermiLab\1-BINS.csv'
ff = pd.read_csv(file_pathrea2)

freq1=ff['Frequency'][0:100]
    
binsx=np.linspace(0, 100, 101)[:-1]
```

```python
# make the list of computed (simulated) elections
elections=[]
for i in range(len(dfsigma)):
    election=[]
    for j in range(len(dfsigma.T[0])):
        
        election.append(dfsigma.T[i][j])

    elections.append(election)
```

```python
# Show the results for 5% bin size data
plt.figure(figsize=(12, 6))
bins = np.linspace(0, 100, 21)[:-1]

for i in range(len(elections)):
    plt.plot(bins,elections[i], linestyle='-', color='blue')
    plt.plot(percent, freq, color='red')
    
plt.xlabel('Percentage of Votes for Georgian Dream in Precincts (%)')
plt.ylabel('Number of Election Precincts')
plt.title('Distribution of Georgian Dream Vote Percentage Across election Precincts, binsize=5%')
plt.grid(True)
plt.savefig('5%bin.png')
plt.show()
```
